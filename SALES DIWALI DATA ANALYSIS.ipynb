```python


# Count plot for a categorical column, e.g., 'Gender'
sns.countplot(data=df, x='Gender')
plt.title('Count of Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# 📊 1. Gender Analysis

# Count Plot - Gender
sns.countplot(data=df, x='Gender')
plt.title('Count of Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Bar Plot - Total Sales by Gender
gender_sales = df.groupby('Gender')['Amount'].sum().reset_index()
sns.barplot(data=gender_sales, x='Gender', y='Amount')
plt.title('Total Purchase Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.show()

# Pie Chart - Gender Distribution
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()

# 📊 2. State Analysis

# Count Plot - Customers by State
plt.figure(figsize=(15,5))
sns.countplot(data=df, x='State', order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Customer Count by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Bar Plot - Total Sales by State
state_sales = df.groupby('State')['Amount'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(15,5))
sns.barplot(data=state_sales, x='State', y='Amount')
plt.xticks(rotation=90)
plt.title('Total Sales by State')
plt.xlabel('State')
plt.ylabel('Total Amount')
plt.tight_layout()
plt.show()

# Pie Chart - Top 5 States by Customer Count
top_states = df['State'].value_counts().head(5)
plt.pie(top_states, labels=top_states.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 States by Customer Count')
plt.axis('equal')
plt.show()
```
